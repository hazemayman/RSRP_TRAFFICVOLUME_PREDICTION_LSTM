```python


def groupSampling(group , window = 24 ):
    values = group.values
    numberOfSamples = len(values) - window
    samples = []
    targets = []
    for i in range(numberOfSamples):
        sample = values[i:window+i]
        target = np.array([values[window+i][1] ,values[window+i][5] ])
        # print(i , '-> ', window+i+1 , ' -> ' , sample.shape)
        samples.append(sample)
        targets.append(target)
    samples = np.reshape(samples , (numberOfSamples , window , -1))
    targets = np.reshape(targets , (numberOfSamples , 1 , -1))
    return (samples,targets)
```

```python
DataSet = None
targets = None
condition = True
for group in groups:
    (sample,target) = groupSampling(group)
    if(condition):
        DataSet = sample
        targets = target
        condition = False
    else:
        DataSet = np.concatenate((DataSet,sample))
        targets = np.concatenate((targets,target))

print(DataSet.shape)
print(targets.shape)
```

```python
from sklearn.utils import shuffle
DataSet_shuffled, targets_shuffled = shuffle(DataSet,targets)
DataSet_shuffled, targets_shuffled = shuffle(DataSet_shuffled,targets_shuffled)

```

```python
import torch
from torch.utils.data import TensorDataset, DataLoader
```

```python
split_frac = 0.85

## split data into training, validation, and test data (features and labels, x and y)

split_idx = int(len(DataSet)*split_frac)
train_x, remaining_x = DataSet_shuffled[:split_idx], DataSet[split_idx:]
train_y, remaining_y = targets_shuffled[:split_idx], targets[split_idx:]

test_idx = int(len(remaining_x)*0.5)
val_x, test_x = remaining_x[:test_idx], remaining_x[test_idx:]
val_y, test_y = remaining_y[:test_idx], remaining_y[test_idx:]

## print out the shapes of your resultant feature data
print("\t\t\tFeature Shapes:")
print("Train set: \t\t{}".format(train_x.shape), 
      "\nValidation set: \t{}".format(val_x.shape),
      "\nTest set: \t\t{}".format(test_x.shape))
```

```python
# create Tensor datasets
train_data = TensorDataset(torch.from_numpy(train_x), torch.from_numpy(train_y))
valid_data = TensorDataset(torch.from_numpy(val_x), torch.from_numpy(val_y))
test_data = TensorDataset(torch.from_numpy(test_x), torch.from_numpy(test_y))

# dataloaders
batch_size = 5

# make sure the SHUFFLE your training data
train_loader = DataLoader(train_data, shuffle=True, batch_size=batch_size)
valid_loader = DataLoader(valid_data, shuffle=True, batch_size=batch_size)
test_loader = DataLoader(test_data, shuffle=True, batch_size=batch_size)
```

```python
# First checking if GPU is available
train_on_gpu=torch.cuda.is_available()

if(train_on_gpu):
    print('Training on GPU.')
else:
    print('No GPU available, training on CPU.')
```

```python
import torch.nn as nn

class SentimentRNN(nn.Module):
    """
    The RNN model that will be used to perform Sentiment analysis.
    """

    def __init__(self, output_size, Feature_dim, hidden_dim, n_layers, drop_prob=0):
        """
        Initialize the model by setting up the layers.
        """
        super(SentimentRNN, self).__init__()

        self.output_size = output_size
        self.n_layers = n_layers
        self.hidden_dim = hidden_dim
        
        # embedding and LSTM layers
        self.lstm = nn.LSTM(Feature_dim, hidden_dim, n_layers, 
                            dropout=drop_prob, batch_first=True)
        
        # dropout layer
        self.dropout = nn.Dropout(drop_prob)
        
        # linear and sigmoid layers
        self.fc = nn.Linear(hidden_dim, output_size)
        

    def forward(self, x, hidden):
        """
        Perform a forward pass of our model on some input and hidden state.
        """
        batch_size = x.size(0)
        
        # embeddings and lstm_out

        # x = x.long()
        lstm_out, hidden = self.lstm(x, hidden)
        lstm_out = lstm_out[:, -1, :] # getting the last time step output
        
        # dropout and fully-connected layer
        out = self.dropout(lstm_out)
        out = self.fc(out)
        # sigmoid function
        
        # return last sigmoid output and hidden state
        return out, hidden
    
    
    def init_hidden(self, batch_size):
        ''' Initializes hidden state '''
        # Create two new tensors with sizes n_layers x batch_size x hidden_dim,
        # initialized to zero, for hidden state and cell state of LSTM
        weight = next(self.parameters()).data
        
        if (train_on_gpu):
            hidden = (weight.new(self.n_layers, batch_size, self.hidden_dim).zero_().cuda(),
                  weight.new(self.n_layers, batch_size, self.hidden_dim).zero_().cuda())
        else:
            hidden = (weight.new(self.n_layers, batch_size, self.hidden_dim).zero_(),
                      weight.new(self.n_layers, batch_size, self.hidden_dim).zero_())
        
        return hidden
        
```

```python
# Instantiate the model w/ hyperparams
output_size = 2
hidden_dim = 512
n_layers = 3
Feature_dim = 9

net = SentimentRNN(output_size, Feature_dim, hidden_dim, n_layers)

print(net)
```

```python
# loss and optimization functions
lr=0.0001

criterion = nn.MSELoss()
optimizer = torch.optim.Adam(net.parameters(), lr=lr)
```

```python
# training params

epochs = 400 # 3-4 is approx where I noticed the validation loss stop decreasing

counter = 0
print_every = 100
clip=5 # gradient clipping

# move model to GPU, if available
if(train_on_gpu):
    net.cuda()

net.train()
# train for some number of epochs
for e in range(epochs):
    # initialize hidden state
    h = net.init_hidden(batch_size)
    # batch loop
    for inputs, labels in train_loader:
        counter += 1

        if(train_on_gpu):
            inputs, labels = inputs.cuda(), labels.cuda()
            inputs = inputs.float()

        # Creating new variables for the hidden state, otherwise
        # we'd backprop through the entire training history
        h = tuple([each.data for each in h])
        # zero accumulated gradients

        net.zero_grad()

        # get the output from the model
        if(inputs.shape[0] != batch_size):
            continue
        output, h = net(inputs, h)
       
        # calculate the loss and perform backprop

        loss = criterion(output.squeeze(), labels.squeeze().float())
        loss.backward()
        
        # `clip_grad_norm` helps prevent the exploding gradient problem in RNNs / LSTMs.
        nn.utils.clip_grad_norm_(net.parameters(), clip)
        optimizer.step()

        # loss stats
        if counter % print_every == 0:
            # Get validation loss
            val_h = net.init_hidden(batch_size)
            val_losses = []
            net.eval()
            for inputs, labels in valid_loader:

                # Creating new variables for the hidden state, otherwise
                # we'd backprop through the entire training history
                val_h = tuple([each.data for each in val_h])

                if(train_on_gpu):
                    inputs, labels = inputs.cuda(), labels.cuda()
                    inputs = inputs.float()
                    
                if(inputs.shape[0] != batch_size):
                    continue
                output, val_h = net(inputs, val_h)

                val_loss = criterion(output.squeeze(), labels.squeeze().float())

                val_losses.append(val_loss.item())

            net.train()
            print("Epoch: {}/{}...".format(e+1, epochs),
                  "Step: {}...".format(counter),
                  "Loss: {:.6f}...".format(loss.item()),
                  "Val Loss: {:.6f}".format(np.mean(val_losses)))
```
